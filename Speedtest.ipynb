```python
import numpy as np
import time
#speed test
size = 1_000_000

l1 = list(range(size))
l2 = list(range(size))
start =time.time()
# python zip uses
add = [ x+y for x,y in zip(l1,l2)]
end = time.time()
print("total time taken by python addition:", end - start)

# using numpy

l3 = np.array(list(range(size)))
l4 = np.array(list(range(size)))
startNp =time.time()
add = l3+l4
endNp = time.time()
print("total time taken by numpy:", endNp - startNp)

```

```python
# python list vs numpy
size =1000
l1 = list(range(size))
start = time.time()
list_sq = [x**2 for x in l1 ]
end = time.time()
print("python list operation result:", list_sq)
print("total time taken by list op:", end - start)
```

```python
size =1000
l1 = np.array(list(range(size)))
start = time.time()
np_sq = l1**2
end = time.time()
print("np result:", np_sq)
print("total time taken by np op:", end - start)
```
